## Predictions

Lets come back to the Titanic example we saw before in the introduction to data preprocessing before. 

We would like to predict who died or not based on some charateristics ( age, sex, etc...), well... this is creppy, but will be memorable... every data scientist in the world has seen this example. 

#### QUESTION : If you where to code a python code manually to try to predict which person survived or not how would you do it ?





I suppose you would try to identify any patterns and calculate the probability of dying for each conditions ? say age>45 died less, sex == F died less etc ? Right ? 

You would then write a code in python which looks like this:

````python
import pandas as pd

df  = pd.read_csv('../../Datasets/titanic.csv')
df = df[df['Age'].notna() & df['Sex'].notna() ]
````

````python
def super_AI_model(age,sex):
    
    if age > 45:
        
        if sex == "F":
            return "20 % chances of dying"
        else:
            return "35 % chances of dying"
        
    else:
        if sex == "F":
            return "50 % chances of dying"
        else:
            return "85 % chances of dying"
````

This can be represented as a decision tree

## Decision Trees

How can we make this automatic ???

````python
# convert to categorical the sex columns example F=0, M=1 
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
df["Sex_C"] =ord_enc.fit_transform(df[["Sex"]])


# train the decision tree

from sklearn import tree,model_selection
train, test = model_selection.train_test_split(df, test_size=20, shuffle=True)
clf = tree.DecisionTreeClassifier()
X_columns = ["Age", "Sex_C"]
Y_columns = ["Survived"]

X= train[X_columns]
Y = train[Y_columns]
clf = clf.fit(X, Y)
````

````python
import graphviz
from sklearn.tree import export_graphviz

# Visualize the tree
from IPython.display import display
display(graphviz.Source(export_graphviz(clf,class_names=["Age", "Sex"],filled=True)))
````

# Random Forest

Random forests (Breiman, 2001) 

The random forest algorithm or "Random Forest" for short is a Machine Learning Method/ Model and belong to the sub-category of meta-estimator. It consist of an ensemble of decisions trees which perform a prediction. This prediction result of the mean of individual predictions of each tree. To be efficient, "Random Forests" combine two sampling methods for training the "Bagging" and the selection of random feature sub-space "random feature selection". The main objective is to minimize overfitting while maximizing accuracy.

<img src="./tree.jpg" />

### Two key concepts , bagging and the selection of random sub space

##### Problem

One problem that we need to solve for the "Ensemble" method to work is the one of "over correlation" between estimators. ie. in the case of Random Forest, if the all the trees are identical, the tree forest is equivalent to a single decision tree. It is necessary to obtain different trees ( who have a low correlation). This problem is accentuated in the big data context , where we have access to big datasets with an high cardinality.
Breiman in his paper introduces the concept of correlation between trees, claiming that the objective is to maximize 
the strengh of each tree while minimizing the correlation between these. To do this, two strategies are put into place:


## 1. Bagging

Considering a $ D $ training dataset of size n, the bagging process generates $ m $ new $ D_i $ datasets, each of size $ n '$, by sampling $ D $ uniformly and with replacement . 

By sampling with replacement, some observations can be repeated many times within each $ D_i $.
By convention, if $ n $ is large we take $ n = n '$ and therefore the dataset $ D_i $ contains $ (1 - \frac {1} {e}) $ $ \approx 63.2 %  $ % single sample, the rest being duplicates. Then in the case of the Random Forest, $ m $ decision trees are trained using the above $ m $ datasets and combined by averaging the output (for regressions) or by performing a majority vote (for classification) .







## 2. Random features selection

The principle is as follows, for each tree making up the "forest" and for each node of these trees we choose a random number $ j $ of dimensions (features). This quite intuitively forces the trees to be different
